```python
#86. Character ASCII Value

#Write a Python program to get the ASCII value of a character. 
def ASCII():
    a = input("Enter a character:" )
    if len(a) > 1:
        print("Please enter only one character")
        return
    b = ord(a)
    print("ASCII value:", b)
        
ASCII()
```
