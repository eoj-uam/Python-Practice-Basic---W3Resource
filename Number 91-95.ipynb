```python
#92 Special Characters in String

#Write a Python program to define a string containing special characters in various forms. 

# Defining strings with special characters

# Using escape sequences
str1 = "This is a newline:\nAnd this is a tab:\tTabbed text."

# Escaping quotes
str2 = "He said, \"Python is fun!\""
str3 = 'It\'s a great day!'

# Backslash character
str4 = "This path uses backslashes: C:\\Users\\Enock\\Documents"

# Unicode characters
str5 = "Unicode examples: \u2600 \u03A9 \u2764"  # Sun, Omega, Heart

# Raw string (ignores escape sequences)
str6 = r"This is a raw string: \n \t \\ stays as it is."

# Print all strings
print(str1)
print(str2)
print(str3)
print(str4)
print(str5)
print(str6)
```

```python
#94. String Bytes to Integers

#Write a Python program to convert the bytes in a given string to a list of integers. 

my_text = "A new String"

# Convert to bytes
byte_data = my_text.encode()   # default UTF-8 encoding

# Convert each byte to integer
int_list = list(byte_data)

print("Original string:", my_text)
print("Bytes:", byte_data)
print("List of integers:", int_list)
```

```python
#95. Check if String is Numeric

#Write a Python program to check whether a string is numeric.

my_str = "This is a new string without numeric characters"

def Ifnum(s):
    if s.isnumeric():
        print("the string is numeric")
    else:
        print("The string is not numeric")

Ifnum(my_str)
```
