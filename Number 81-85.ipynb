```python
#82. Sum of Container Items

#Write a Python program to calculate the sum of all items of a container (tuple, list, set, dictionary).
#tuple
tup1 = (10, 21, 65, 55, 885, 454)
sum1 = sum(tup1)
print("Tuple sum:", {sum1})

#list
lst = [5, 15, 25]
total = sum(lst)
print("List sum:", total)

#set
st = {23, 34, 35, 45, 35}
sumset = sum(st)
print("Set sum:", sumset)

#dictionary
d = {"a": 10, "b": 20, "c": 30}

total_values = sum(d.values())
print("Dictionary sum:", total_values)
```

```python
#83. List Greater-Than Test

#Write a Python program to test whether all numbers in a list are greater than a certain number.

lst = [96,48,69, 22, 77, 39, 21, 41, 64, 16, 14, 75, 31, 98]
num1 = int(input("Enter a number to check:"))

if all(i > num1 for i in lst):
    print(f'All number are greater than {num1}')
else:
    print(f'Not all numbers are greater than {num1}')

greater_nums = [i for i in lst if i > num1]
print(f'Numbers greater than {num1} are {greater_nums}.')
```
